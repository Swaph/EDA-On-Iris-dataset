#Exploratory Data Analysis (EDA) on the Iris Dataset

##Introduction
The Iris dataset is one of the most famous datasets used in data science and machine learning. It contains measurements of sepal length, sepal width, petal length, and petal width for 150 iris flowers, categorized into three different species: Setosa, Versicolor, and Virginica. This dataset is often used as a beginner's dataset for pattern recognition and classification techniques.

In this notebook, I will perform an Exploratory Data Analysis (EDA) on the Iris dataset to gain insights into its structure and the relationships between its features. EDA is a crucial step in the data analysis process, as it helps us understand the data, detect anomalies, and identify patterns that can inform further analysis or model building.

###Step-by-step process



####Importing Libraries
These libraries are Pandas for data handling, Matplotlib for plotting, and Seaborn for both the dataset and statistical visualizations.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set seaborn style for better visualization
sns.set(style="whitegrid")
```

####Loading the Dataset

This dataset has be directly loaded from the Seaborn library.

```python
# Load the dataset from seaborn library
iris = sns.load_dataset('iris')

#Show first 5 rows to see what the dataset looks like
iris.head()
```

####Initial Exploration

In this step, I will perform a preliminary exploration to understand the dataset's structure and basic statistics. This includes checking the data types, missing values, and summary statistics.

```python
#Get the shape of the dataset.
iris.shape
```

```python
#Display basic information about the iris dataset
iris.info()
```

```python
#Summary statistics of the iris dataset
iris.describe()
```

```python
#Check for Missing Values
iris.isnull().sum()
```

```python
#Check for Duplicates
iris.drop_duplicates(subset ="species",)
```

```python
#Check if the dataset is balanced
iris.value_counts("species")
```

####Univariate Analysis

First, I will analyze each feature individually to understand its distribution. This involves creating histograms and box plots for each feature.

```python
# Histograms for each feature (sepal length & width and petal length & width)
iris.hist(figsize=(10, 8), bins=20)
plt.show()

# Box plots for each feature too
plt.figure(figsize=(10, 8))
sns.boxplot(data=iris)
plt.show()
```

####Bivariate Analysis

Next, I will try exploring the relationships between pairs of features. I will use pair plots to visualize these relationships and a heatmap to see the correlation between the features.

```python
# Pairplot to visualize relationships between features
sns.pairplot(iris, hue='species', markers=["o", "s", "D"])
plt.show()

# Correlation heatmap, excluding the 'species' column
plt.figure(figsize=(10, 8))
sns.heatmap(iris.drop('species', axis=1).corr(), annot=True, cmap='coolwarm')
plt.show()
```

####Multivariate Analysis

Lastly, I will examine relationships among multiple features simultaneously. This helps us understand how different features interact with each other across different species.

```python
# Pairwise relationships with species hue
sns.pairplot(iris, kind="scatter", hue="species", markers=["o", "s", "D"])
plt.show()

# Violin plots for each feature grouped by species
plt.figure(figsize=(15, 10))
for i, feature in enumerate(iris.columns[:-1]):
    plt.subplot(2, 2, i + 1)
    sns.violinplot(x='species', y=feature, data=iris, hue='species', palette='muted', legend=False)
    plt.title(feature)
plt.tight_layout()
plt.show()
```
